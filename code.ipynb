```python
data_list = data.values.tolist()
```

```python
print(data_list)
```

```python
print (np.ndim(data_list))
```

```python
clean_text=[]

for words in data_list :
    clean_text2=[]
    for i in words:
        clean_text2.append(str.lower(i)) 
        clean_text.append(clean_text2)
```

```python
print(clean_text)
```

```python
from nltk import word_tokenize as wtk
```

```python
import re
```

```python
clean_text2=[[j] for i in clean_text for j in i ]
```

```python
clean_text2
```

```python
print(np.shape(clean_text))
```

```python
from nltk.corpus import stopwords as sp
import re
```

```python
clean_text3=[[re.sub(r'[^\w\s]',"",j)] for i in clean_text for j in i if re.sub(r'[^\w\s]',"",j)!=""]
print(clean_text3)
```

```python
clean_text_=[wtk(sublist) for i in clean_text3 for sublist in i ]
```

```python
print(clean_text_)
```

```python
compare=set(sp.words('English'))
print(compare)
```

```python
aft_stop =  [
        [word for word in sentence if word not in compare]  
        for sentence in clean_text_
    ]


print(aft_stop)
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
ps=PorterStemmer()
wl=WordNetLemmatizer()
```

```python
aft_stem= [
             [ps.stem(word) for word in sentence]
             for sentence in aft_stop
           ]
```

```python
aft_stem
```

```python
aft_lem= [
    [wl.lemmatize(i)  for i in aft]
    for aft in aft_stem
  ]
```

```python
df= pd.DataFrame(aft_lem)
```

```python
sentences = [" ".join(inner_list) for inner_list in aft_lem]
```

```python
print(sentences)
```

```python
df2 = pd.DataFrame(sentences, columns=['Tweets'])
df2.to_csv('preprocess4.csv')
```

```python

import matplotlib.pyplot as plt
```

```python
print(df2)
```

```python
from textblob import TextBlob

for index, row in df2.iterrows():  
    tweet_text = row['Tweets'] 
    analysis = TextBlob(tweet_text)
    
    polarity = analysis.sentiment.polarity

    if polarity > 0:
        sentiment = "Positive"
    elif polarity < 0:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    
    df2.at[index, 'Sentiment'] = sentiment
    print(f"Tweet {index + 1} is {sentiment}: {tweet_text}")

# df2.to_csv('completed.csv')  
       
```

```python
sentiment_counts = df2['Sentiment'].value_counts()
print(sentiment_counts)
```
